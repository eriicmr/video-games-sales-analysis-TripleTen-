# Step 5. Testing Hypotheses 

**1.** Formulating the Hypoteses (Platforms):

First we got to define the hypoteses
- NULL Hypoteses: The mean rating from XBOX One and PC users are equal.
- Alternative Hypoteses: The mean rating from XBOX One and PC uers are different.

**2.** Choosing a Statistical Test:
- Since we are comparing the means between two groups, we can use T-Test for independent samples.

```python
#Before start testing, we need to import scipy library
from scipy.stats import ttest_ind, levene
```

```python
#Filtering data from both platforms XBOX and PC
xbox_user_score = df[df['platform'] == 'XOne']['user_score'].dropna()
pc_user_score = df[df['platform'] == 'PC']['user_score'].dropna()

stat_levene, p_levene = levene(xbox_user_score, pc_user_score)

#Evaluating similarity between variants
if p_levene > 0.05:
    equal_var = True
    print("Variants are equal. (p_value levene's test: {:.4f})".format(p_levene))
else:
    equal_var = False
    print("Variants are different. (p_value levene's test: {:.4f})".format(p_levene))


#Starting the test
t_stat, p_value = ttest_ind(xbox_user_score, pc_user_score, equal_var=equal_var)

print('Statistics from the T-Test:', t_stat)
print('P-Value is:', p_value)


alpha = 0.05
if p_value < alpha:
    print('We reject null hypotesis. Medians are significantly different')
else:
    print('We do not reject the null hypothesis. There is not eonugh evidence showing mediand are significantly different ')
```

# Interpreting the T-Test Results

* The t-statistic is negative, which indicates that the mean user score for Xbox One is lower than that for PC.
* The magnitude of this value shows how far the observed difference is from zero in terms of standard error.

`The p-value here is 1.39 × 10 ^ -5 which is extremely small (essentially 0.0000139).`

**Decision Rule**
* The usual threshold/alpha value (α) for significance is 0.05.
* Since the p-value is much smaller than 0.05, we reject the null hypothesis.


**Conclusion**
* The results suggest that the average user scores for Xbox One and PC are statistically different.
* This means the hypothesis that their means are the same is unlikely to be true.

2. Formulating Hypoteses (Genres):

2.1. Choosing the Statistical Test:

```python
#Filtering data of both selected genres.
rating_action = df[df['genre'] == 'Action']['user_score'].dropna()
rating_sport = df[df['genre'] == 'Sports']['user_score'].dropna()

#Creating the test.
t_stat, p_value_levene = levene(rating_action, rating_sport)

equal_var = p_value_levene > 0.05

t_stat, p_value = ttest_ind (rating_action, rating_sport, equal_var=equal_var)

print('T-Test result:', t_stat)
print('P-value:', p_value)

alpha = 0.05
if p_value < alpha:
    print('We reject the null hypotesis. Medians are statistically different.')
else:
    print('We do not reject the null hypothesis. There is not enough evidence to show medians are different.')
```
